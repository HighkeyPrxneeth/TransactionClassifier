```python
import pandas as pd

names = pd.read_csv('business_names_new.csv', index_col=0)
names
```

```python
categories = pd.read_csv('categories.csv')
categories
```

```python
def get_category_by_id(cat_id):
    row = categories[categories['category_id'] == cat_id]
    if not row.empty:
        return row.iloc[0]['category_label']
    return None

names['category_label'] = names['fsq_category_ids'].apply(get_category_by_id)
names
```

```python
names = names[names["Count"] >= 5]
```

```python
names.to_csv('business_names_with_categories.csv', index=False)
```
